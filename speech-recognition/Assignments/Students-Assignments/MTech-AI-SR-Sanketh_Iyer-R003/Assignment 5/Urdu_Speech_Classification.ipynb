```python
from sklearn.preprocessing import minmax_scale
def normalize(x, axis=0):
    return minmax_scale(x, axis=axis)
```

```python
def AudioPlot(audio,name):
    fig,axis=plt.subplots(7,figsize=(18,18))
    plt.subplots_adjust(hspace=0.8)
    x , sr = librosa.load(audio)
    axis[0].title.set_text(name+" - Waveplot")
    librosa.display.waveshow(x, sr=sr,ax=axis[0])
    axis[1].title.set_text(name+" - Spectrogram")
    X = librosa.stft(x)
    Xdb = librosa.amplitude_to_db(abs(X))
    librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log',ax=axis[1])
    axis[2].title.set_text(name+" - Zero Crossing Rate")
    zero_crossings = librosa.zero_crossings(x)
    zcrs = librosa.feature.zero_crossing_rate(x)
    axis[2].plot(zcrs[0])
    axis[3].title.set_text(name+" - Spectral Centroid")
    spectral_centroids = librosa.feature.spectral_centroid(x, sr=sr)[0]
    frames = range(len(spectral_centroids))
    t = librosa.frames_to_time(frames)
    librosa.display.waveshow(x, sr=sr, alpha=0.4,ax=axis[3])
    axis[3].plot(t, normalize(spectral_centroids), color='r')
    axis[4].title.set_text(name+" - Spectral Roll Off")
    spectral_rolloff = librosa.feature.spectral_rolloff(x, sr=sr, roll_percent=0.01)[0]
    frames = range(len(spectral_rolloff))
    t = librosa.frames_to_time(frames)
    librosa.display.waveshow(x, sr=sr, alpha=0.4,ax=axis[4])
    axis[4].plot(t, normalize(spectral_rolloff), color='r')
    axis[5].title.set_text(name+" - MFCC")
    mfccs = librosa.feature.mfcc(y=x, sr=sr)
    librosa.display.specshow(mfccs, sr=sr, x_axis='time',ax=axis[5])
    axis[6].title.set_text(name+" - Mel Spectrogram")
    mel_spect = librosa.feature.melspectrogram(y=x, sr=sr, n_mels=128,fmax=8000)
    mel_spect = librosa.power_to_db(mel_spect, ref=np.max)
    librosa.display.specshow(mel_spect, y_axis='mel', fmax=8000, x_axis='time',ax=axis[6])
```

### MFCCs

```python
import sklearn
```

```python
def create_mfcc_features(path):
    mfccs = []
    try:
        x , sr = librosa.load(path)
        mfccs = librosa.feature.mfcc(y=x, sr=sr, n_mfcc=64)
        mfccs = np.mean(mfccs,axis=1)

    except:
        print('Error reading audio')
    return mfccs
```

```python
X_df = pd.DataFrame(df['Path'].apply(lambda x: create_mfcc_features(x)).tolist())
X_df.head()
```

## 1D CNN

```python
X_train_cnn = X_train.values.reshape(-1,64,1)
X_test_cnn = X_test.values.reshape(-1,64,1)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Conv1D, MaxPooling1D, Dense, Flatten
```

```python
model = Sequential([
    InputLayer(input_shape=(64, 1)),
    Conv1D(64, 11, activation='relu', padding = "same"),
    MaxPooling1D(3),
    Conv1D(64, 9, activation='relu', padding = "same"),
    MaxPooling1D(3),
    Conv1D(64, 7, activation='relu', padding = "same"),
    MaxPooling1D(3),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(64, activation='relu'),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(15, activation='softmax')
])
model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train_cnn, y_train, epochs=100, validation_data=(X_test_cnn,y_test))
```

```python
model.evaluate(X_test_cnn,y_test)
```

```python
plt.figure(figsize=(15,8))
plt.plot(history.history['loss'])
plt.xlabel('EPOCH')
plt.ylabel('LOSS')
plt.title('MODEL LOSS')
plt.legend(['train'])
plt.show()

plt.figure(figsize=(15,8))
plt.plot(history.history['accuracy'])
plt.xlabel('EPOCH')
plt.ylabel('ACCURACY')
plt.title('MODEL ACCURACY')
plt.legend(['train'])
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.plot(history.history['val_loss'])
plt.xlabel('EPOCH')
plt.ylabel('LOSS')
plt.title('MODEL LOSS')
plt.legend(['test'])
plt.show()

plt.figure(figsize=(15,8))
plt.plot(history.history['val_accuracy'])
plt.xlabel('EPOCH')
plt.ylabel('ACCURACY')
plt.title('MODEL ACCURACY')
plt.legend(['test'])
plt.show()
```

```python
cnn_predict=model.predict(X_test_cnn)
cnn_predict = np.argmax(cnn_predict,axis=1)
print(accuracy_score(cnn_predict,y_test))
print(classification_report(cnn_predict,y_test,zero_division=0))
```
